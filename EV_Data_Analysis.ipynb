```python
#Electrical Vehicles Dataset for Data Analysis and Visualization

#The data contains the following fields:
#1.Vehicle_ID: Unique identifier for each vehicle record.
#2.Manufacturer: Vehicle brand or OEM.	
#3.Model: Specific model name/variant.	
#4.Year: Model year associated with the record.
#5.Battery_Type: Battery chemistry/technology used.	
#6.Battery_Capacity_kWh: Nominal battery capacity in kilowatt-hours.	
#7.Range_km: Claimed driving range on a full charge (kilometers).
#8.Charging_Type: Dominant charging interface or feature.
#9.Charge_Time_hr: Approximate time to charge (hours), context varies by charging method.	
#10.Price_US: Indicative vehicle price in US dollars.
#11.Color: Primary exterior color or finish.	
#12.Country_of_Manufacture: Country where the vehicle is manufactured/assembled.
#13.Autonomous_Level: Automation capability level (e.g., 0–5), may include decimals for sub-levels.	
#14.CO2_Emissions_g_per_km: Tailpipe CO2 emissions in g/km (often 0 for BEVs); may be blank for missing data.	
#15.Safety_Rating: Safety rating (scale not explicitly defined; often 1–5).
#16.Units_Sold_2024: Units sold in 2024 for this model/variant (as recorded).
#17.Warranty_Years: Warranty duration in years.


import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
ev = pd.read_csv("electric_vehicles_dataset.csv")
ev.head()
```

```python
obj = (ev.dtypes == 'object')
object_cols = list(obj[obj].index)

unique_values = {}
for col in object_cols:
    unique_values[col] = ev[col].unique().size
print(unique_values)
```

```python
ev.duplicated().sum()
```

```python
#Data Visualization
#1.Charging Type by Charging Time
plt.figure(figsize=(10,5))
sns.barplot(x='Charging_Type',y='Charge_Time_hr',data=ev,palette='coolwarm')
plt.title('Charging Type by its Time')
plt.ylabel('Charging Time in hr')
plt.xticks(rotation=90)
plt.show
```

```python
#2.Top 10 Countries of Manufacturing
sns.countplot(y='Country_of_Manufacture',data=ev,palette='Set1',order=ev['Country_of_Manufacture'].value_counts().head(10).index)
plt.title('Top 10 Countries of Manufacture')
plt.ylabel('Country')
plt.show()
```

```python
#3.Battery Capacity vs Range
sns.scatterplot(x='Battery_Capacity_kWh',y='Range_km',data=ev,marker='+')
plt.title('Battery vs Range')
plt.xlabel('Battery Capacity in kWh')
plt.ylabel('Range in km')
plt.show
```

```python
#4.Number of Units sold in 2024 by Countries 
plt.figure(figsize=(15,5))
sns.pointplot(x='Country_of_Manufacture',y='Units_Sold_2024',data=ev,color='blue')
plt.title('Countries sold unit in 2024')
plt.xlabel('Countries')
plt.ylabel('Units Sold')
plt.xticks(rotation=90)
plt.show
```

```python
#5.Top 10 colors which is most purchased
sns.countplot(y='Color',data=ev,palette='Set2',order=ev['Color'].value_counts().head(10).index)
plt.title('Top 10 Colors')
plt.xlabel=('Color')
plt.show()
```

```python
#6.Battery Type by Battery Capacity
plt.figure(figsize=(10,5))
sns.barplot(x='Battery_Type',y='Battery_Capacity_kWh',data=ev,palette='colorblind')
plt.title('Battery Type by its Capacity ')
plt.ylabel('capacity in kWh')
plt.xticks(rotation=90)
plt.show
```

```python
#7.Top 10 Manufacturing Brands
sns.countplot(y='Manufacturer',data=ev,order=ev['Manufacturer'].value_counts().head(10).index,palette='Set3')
plt.title('Top 10 Manufacturer')
plt.show()
```

```python
#8.Warranty Years by Safety Rating
sns.catplot(x='Warranty_Years',hue='Safety_Rating',data=ev,kind='count')
plt.show()
```
